# import necessary library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from datetime import datetime
```

# download dataset from yahoo finance

```python
start = datetime(year=2018, month=1, day=1)
finish = datetime(year=2023, month=1, day=1)
msft =  yf.download(tickers='MSFT',start=start, end=finish)
goog =  yf.download(tickers='GOOG',start=start, end=finish)
```

```python
print('MICROSOFT stock')
print(msft.head())
print('-----')
print('GOOGLE stock')
print(goog.head())
```

# visualize stock from since five years ago

```python
plt.figure(figsize=(16,5))
msft['Adj Close'].plot(color='r', legend='microsoft')
goog['Adj Close'].plot(color='b', legend='google')
plt.legend()
plt.title('Stock Price (close) Microsoft VS Google')
plt.show()
```

```python
plt.figure(figsize=(16,5))
msft['Volume'].plot(color='r', legend='microsoft')
goog['Volume'].plot(color='b', legend='google')
plt.legend()
plt.title('Stock Price (volume) Microsoft VS Google')
plt.show()
```
