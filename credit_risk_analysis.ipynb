# import library

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
import matplotlib.pyplot as plt
import seaborn as sns
```

# load dataset

```python
loc_data = input('where is your data? ')
```

```python
dataframe = pd.read_csv(loc_data)
dataframe.head()
```

```python
dataframe = dataframe.drop([dataframe.keys()[0]], axis=1)
dataframe.head(1)
```

# data cleaning

### check data type

```python
df = dataframe
df.info()
```

### clean variable term

```python
df['term'].unique()
```

```python
pd.get_dummies(data=df['term'], drop_first=True).head()
```

```python
df['term_60_months'] = pd.get_dummies(data=df['term'], drop_first=True).values
df = df.drop(['term'], axis=1)
```

I change variable 36 and 60 months into boolean. 0 for 36 months and 1 for 60 months. And I drop variable _original_ term because it no longer needed.

```python
df.tail()
```

### clean variable emp_length

```python
df['emp_length'].unique()
```

```python
df['emp_length_int'] = df['emp_length'].str.replace('\+ years', '')
df['emp_length_int'] = df['emp_length_int'].str.replace('< 1 year', str(0))
df['emp_length_int'] = df['emp_length_int'].str.replace('years', '')
df['emp_length_int'] = df['emp_length_int'].str.replace('year', '')
```

```python
print(df['emp_length_int'].unique())
print(type(df['emp_length_int'][0]))
```

```python
df['emp_length_int'] = pd.to_numeric(df['emp_length_int'])
print(type(df['emp_length_int'][0]))
df = df.drop(['emp_length'], axis=1)
```

First I see how many unique values in emp_length. After that, I change less than 1 year into interger 0, more than 10 years into interger 10, and I treat the rest by erase the word year to make it into interger. Last, I drop _original_ emp_length variable because it no nonger needed.

### change string to datetime
some variables are supposed to be datetime but detected as string.

```python
df['issue_date'] = pd.to_datetime(df['issue_d'], format='%b-%y')
df['earliest_cr_line_date'] = pd.to_datetime(df['earliest_cr_line'], format='%b-%y')
df['last_payment_date'] = pd.to_datetime(df['last_pymnt_d'], format='%b-%y')
df['last_credit_pull_date'] = pd.to_datetime(df['last_credit_pull_d'], format='%b-%y')
```

```python
df = df.drop(['issue_d', 'earliest_cr_line', 'last_pymnt_d', 'last_credit_pull_d'], axis=1)
```
