# import necessary library

```python
import numpy as np
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

# download dataset

```python
df = yf.download('PG', start='1995-1-1')
```

```python
df = df['Adj Close']
```

```python
df.plot(figsize=(16,4))
plt.show()
```

# the simple rate of return

```python
simple_diff = pd.Series((df/df.shift(1)) - 1)
simple_diff.plot(figsize=(16,4))
plt.show()
```

```python
working_day = 250
percentage = 100
simple_rate = (simple_diff.mean()*percentage*working_day).round(3)
print('the simple rate of return is ', simple_rate, '%')
```

# the logarithmic rate of return

```python
log_diff = pd.Series(np.log(df/df.shift(1)))
log_diff.plot(figsize=(16,4))
plt.show()
```

```python
working_day = 250
percentage = 100
log_rate = (log_diff.mean()*percentage*working_day).round(3)
print('the logarithmic rate of return is ', log_rate, '%')
```

# the comparison between both rate of return

```python
plt.figure(figsize=(16,4))
log_diff.plot()
simple_diff.plot()
plt.show()
```

```python
if simple_rate > log_rate:
    print('simple rate is higher than log rate with different ', abs(simple_rate-log_rate).round(3))
elif log_rate > simple_rate:
    print('logarithmic rate is higher than simple rate with different ', abs(log_rate-simple_rate).round(3))
else:
    print('both rate of return methods have same value')
```
